```python
import matplotlib.pyplot as plt
import numpy as np

plt.suptitle("GRAPHS",fontsize=16)

x = np.array([2,8])
y = np.array([0,8])
plt.subplot(3,3,1)
plt.scatter(x,y)
plt.xlabel("xlabel",color="blue")
plt.ylabel("ylabel",color="blue")
plt.tight_layout()

x1=np.array([2,4,5,6,7])
y1=np.array([2,4,2,4,2])
plt.subplot(3,3,2)
plt.plot(x1,y1,marker='*',color="orange")

x2=np.array([2,3,5,5,6,7])
y2=np.array([3,2,3,3,2,3])
plt.plot(x2,y2,linestyle="dotted",color="blue")
plt.tight_layout()

x=np.array([3,4,5,6])
y=np.array([2,4,2,5])
plt.subplot(3,3,3)
plt.plot(x,y,marker="o",mfc="yellow",mec="black")
plt.tight_layout()

x=np.array([2,4,6,8])
y=np.array([2,4,6,8])
plt.subplot(3,3,4)
plt.plot(x,y)
plt.grid()
plt.tight_layout()

x=np.array([2,1,3,7,4,8,4,6])
y=np.array([4,5,9,4,7,3,9,5])
colors=["red","green","blue","yellow","purple","cyan","magenta","orange"]
plt.subplot(3,3,5)
plt.scatter(x,y,c=colors)
plt.tight_layout()

x=np.array([2,5,8])
y=np.array([5,3,7])
plt.subplot(3,3,6)
plt.bar(x,y,color="red",width=0.5)
plt.tight_layout()

x = np.array([35, 25, 20, 20])
mylabels = ["explode", "shadow", "color", "legend"]
explode = [0.1, 0, 0, 1]
colors = ["red", "green", "blue", "yellow"]


plt.figure(figsize=(10, 8))
plt.subplot(3, 3, 7)
plt.pie(x, labels=mylabels, explode=explode, shadow=True, colors=colors, autopct='%1.1f%%')
plt.title("Pie Chart with All Features")
plt.legend(title="Labels", loc="upper right")
plt.tight_layout()
plt.show()
```
